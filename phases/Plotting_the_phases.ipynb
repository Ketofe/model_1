```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def phase_plotter(variable0,variable1,s,label0,label1):
    for j in range(len(variable0)):
        for l in range(len(variable1)):
            if s[j][l]==1:
                plt.scatter(variable0[j],variable1[l],color='yellow' )
            elif s[j][l]==0:
                plt.scatter(variable0[j],variable1[l],color='blue' )
            else:
                plt.scatter(variable0[j],variable1[l],color='green' )
    plt.xlabel(label0,fontsize=23)
    plt.ylabel(label1,fontsize=23)
```

```python
eta_list=np.linspace(0,np.pi,20)
alpha_list=np.linspace(0,4,20)


alpha_eta_r2=np.load("alpha_eta/beta1_pahse_plot_r2.npy")
alpha_eta_r3=np.load("alpha_eta/beta1_pahse_plot_r3.npy")
alpha_eta_r4=np.load("alpha_eta/beta1_pahse_plot_r4.npy")
alpha_eta_r5=np.load("alpha_eta/beta1_pahse_plot_r5.npy")
```

```python
phase_plotter(alpha_list,eta_list,alpha_eta_r2,r'$\alpha$',r'$\eta$')
plt.savefig("Plots/alpha_eta_r2.pdf")
```

```python
#Now critical alphas at eta=0
alpha_critical=np.array([(alpha_list[12]+alpha_list[13])/2,(alpha_list[6]+alpha_list[7])/2,(alpha_list[5]+alpha_list[6])/2,alpha_list[5]])
r_critical=np.array([2,3,4,5])
plt.scatter(r_critical,alpha_critical)
plt.ylabel(r'$\alpha$',fontsize=23)
plt.xlabel(r'$R$',fontsize=23)

from scipy.optimize import curve_fit
def curve_trial(r,p):
    return p/r
paramter,_ = curve_fit(curve_trial,alpha_critical ,r_critical)
rs=np.linspace(2,5,100)
plt.plot(rs,curve_trial(rs,paramter[0]))
print(paramter[0])
plt.savefig("plots/r_alpha_critical_eta0_scatter.pdf")
```
