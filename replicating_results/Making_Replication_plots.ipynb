```python

import numpy as np
import matplotlib.pyplot as plt





```

```python
def order(angle_list):
    return np.sqrt(sum(np.sin(angle_list))**2+sum(np.cos(angle_list))**2 )/len(angle_list)



#For order and fraction of coopeators plot
def order_and_fraction_of_cooperators(theta,s):
          o=[order(k) for k in theta]
          C=[sum(k)/len(k) for k in s]
          plt.plot(C,label='C')
          plt.plot(o,label=r'$\Psi$')
          plt.plot()
          plt.xscale('log')
          plt.legend(fontsize=23)
          plt.xlabel('t',fontsize=23)

#Need to pos[i] etc
def Quiver_plot(pos,theta,s):
       plt.quiver(pos[:,0],pos[:,1],np.cos(theta),np.sin(theta) ,s,clim=[0,1]) 
       plt.xlabel('x',fontsize=23)
       plt.ylabel('y',fontsize=23)
```

```python

#Now making animtation
from matplotlib.animation import FuncAnimation
#https://francescoturci.net/2020/06/19/minimal-vicsek-model-in-python/ was used to make this


def animate(pos,theta,s,number_of_frames):

        fig, ax= plt.subplots(figsize=(6,6))
            

        qv=ax.quiver(pos[0][:,0], pos[0][:,1], np.cos(theta[0]), np.sin(theta[0]), s[0], clim=[0, 1])

        # Animation function
        def animate(i):
            

            qv.set_offsets(pos[i])
            qv.set_UVC(np.cos(theta[i]), np.sin(theta[i]),s[i])
           
            
            
            return qv,

        anim = FuncAnimation(fig,animate,np.arange(1, number_of_frames),interval=1, blit=True)
        plt.show()
        return anim




```

```python

anim=animate(pos_a,theta_a,s_a,1000)
anim.save("a_animate.gif")
```
